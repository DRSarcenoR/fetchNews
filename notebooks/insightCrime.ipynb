## prueba buscador basico

```python
url = 'https://insightcrime.org/es/'
```

```python
nombre = 'bernardo arevalo'
```

```python
nombre2 = '%20'.join(nombre.split())
```

```python
response = requests.get(
    url + '?s=' + nombre2 + '#main', 
    proxies={'http': None, 'https': None}, 
    timeout=60,
    verify=False
)
```

```python
response.status_code
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
with open('../data/insightCrime/prueba_v2.html', 'w') as ic:
    ic.write(soup.prettify())
```
